```python
from bs4 import BeautifulSoup as bs
import requests

page=requests.get("https://www.amazon.in/iPhone-Pro-Max-256GB-Gold/product-reviews/B0BDK63WMS?reviewerType=all_reviews")
page
```

```python
page.content
```

```python
soup = bs(page.content,'html.parser')
print(soup.prettify())
```

```python
name=soup.find_all('span', class_='a-profile-name')
name
```

```python
for i in range(0,len(name)):
    print(name[i])
```

```python
len(name)
```

```python
cust_name = []
for i in range(0, len(name)):
    cust_name.append(name[i].get_text())

for name in cust_name:
    print(name)
    
print('\n')
len(cust_name)
```

```python
cust_name=list(set(cust_name))
for name in cust_name:
    print(name)
    
len(cust_name)
```

```python
title=soup.find_all('span', class_='a-size-base review-title a-text-bold')
title
```

```python
review_title=[]
for i in range(len(title)):
    review_title.append(title[i].get_text())
    
for title in review_title:
    print(title)
```

```python
review_title.append(" ")
```

```python
len(review_title)
```

```python
import pandas as pd
df=pd.DataFrame()
```

```python
df['Customer name']=cust_name
```

```python
df['Review Title']=review_title
```

```python
import xlsxwriter

writer=pd.ExcelWriter("amazon_product.xlsx", engine="xlsxwriter")
df.to_excel(writer, sheet_name='List')
writer.close()
```
