```python
import pandas as pd
```

```python
df=pd.read_csv("dataset.csv")
```

```python
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#linear regression
x=np.array(df.depth)
y=np.array(df.age)
```

```python
n=np.size(x)
mean_x=np.sum(x)/n
mean_y=np.sum(y)/n
```

```python
sum_x_y=np.sum(x*y)-(n*mean_x*mean_y)
sum_x_x=np.sum(x*x)-(n*mean_x*mean_x)
b_1= sum_x_y/sum_x_x
b_0=mean_y-(b_1*mean_x)
print("estimated coefficients:\n b_0:", b_0, "\n b_1=", b_1)
```

```python
plt.scatter(x,y, s=30)
```

```python
y_predicted=b_0 + b_1*x
```

```python
plt.scatter(x,y, s=30)
plt.plot(x, y_predicted)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Estimating y value
xi=600
yi=b_0+b_1*xi
yi
```

```python
#residual plot
y_pred=b_0+b_1*x
residual =y-y_pred
plt.scatter(y_pred, residual)
y_0 =[0,0,0,0,0,0,0]
plt.plot(y_pred,y_0)
```
