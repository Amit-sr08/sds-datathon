```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
from scipy.stats import norm
```

```python
n=100
x=np.random.randn(n)
x
```

```python
data=sorted(x)
data
```

```python
q=[(i-0.5)/len(data) for i in range(1,len(data)+1)]
#q=[(i-0.5)/len(data) for i in range(1,len(data)+1)]
q
```

```python
z=zscore(q)
z
```

```python
plt.plot(data,z,'ro',data,data)#'bo' is for blue color and 'ro' is for red color
plt.xlabel('Observed values')
plt.ylabel('z score')
plt.show()
```

```python
xq=[norm.ppf(i,np.mean(data),np.std(data)) for i in q]
#xq=[norm.ppf(i,np.mean(data),np.std(data)) for i in q]
xq
```

```python
plt.plot(data,xq,'ro',data,data)
plt.xlabel('observed values')
plt.ylabel('z score')
plt.show()
```

```python
x1=[3.35,4.79,5.96,7.89,3.01]
x2=[3.325,3.654,5.466]

def npp(x):
    data=sorted(x)
    q1=[(i-0.5)/len(data) for i in range(1,len(data)+1)]
    xq1=[norm.ppf(i,np.mean(data),np.std(data)) for i in q1]
    plt.plot(data,xq1,'ro',data,data)
    plt.show()

npp(x1)
npp(x2)
```
